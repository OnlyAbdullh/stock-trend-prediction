```python
gap_counts = gaps_cleaned.groupby("date").size()
data_counts = df_cleaned.groupby("date").size()
gap_counts = (
    gaps_cleaned.groupby("date")
    .size()
    .rename("number_of_missed_records")
    .reset_index()
)

data_counts = (
    df_cleaned.groupby("date")
    .size()
    .rename("number_of_tickers_at_date")
    .reset_index()
)
gap_vs_data = pd.merge_asof(
    gap_counts.sort_values("date"),
    data_counts.sort_values("date"),
    on="date",
    direction="nearest"
)
```

```python
GLOBAL_THRESHOLD = 0.8
import  numpy as np
gap_vs_data["gap_ratio"] = (
    gap_vs_data["number_of_missed_records"] / gap_vs_data["number_of_tickers_at_date"]
)

gap_vs_data["gap_type"] = np.where(
    gap_vs_data["gap_ratio"] >= GLOBAL_THRESHOLD,
    "global",
    "local"
)
```
